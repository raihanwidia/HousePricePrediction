# Replace Outliers of Selected Features`

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

X= CombinedDataHouseData[used_cor['features']]


# 1. Select numeric columns
numeric_cols = X.select_dtypes(include='number').columns

# 2. Copy original data for comparison
X_original = X.copy()

# 3. Function to replace outliers using IQR capping
def cap_outliers(df):
    df_capped = df.copy()
    for col in df_capped.columns:
        Q1 = df_capped[col].quantile(0.25)
        Q3 = df_capped[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df_capped[col] = df_capped[col].clip(lower=lower, upper=upper)
    return df_capped

# 4. Replace outliers
X_capped = cap_outliers(X[numeric_cols])

# 5. Visualize with histograms (Before vs After)
import math

num_cols = len(numeric_cols)
rows = math.ceil(num_cols)

for col in numeric_cols:
    plt.figure(figsize=(12, 4))

    # Histogram before
    plt.subplot(1, 2, 1)
    sns.histplot(X_original[col], kde=True, color="skyblue")
    plt.title(f"{col} - Before")

    # Histogram after
    plt.subplot(1, 2, 2)
    sns.histplot(X_capped[col], kde=True, color="lightgreen")
    plt.title(f"{col} - After")

    plt.tight_layout()
    plt.show()
```

```python
X= CombinedDataHouseData[used_cor['features']]

# Get numeric columns from your DataFrame
numeric_cols = X.select_dtypes(include='number').columns

# Loop through each numeric column and check for outliers
for col in numeric_cols:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = X[(X[col] < lower_bound) | (X[col] > upper_bound)]

    if not outliers.empty:
        print(f"\nOutliers detected in '{col}':")
        print(outliers[[col]])


for col in numeric_cols:
    Q1 = X[col].quantile(0.25)
    Q3 = X[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Replace outliers
    X[col] = X[col].clip(lower=lower, upper=upper)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import xgboost as xgb


LR = LinearRegression()
RF = RandomForestRegressor(random_state=42)
Ada = AdaBoostRegressor(random_state=42)
Xgb = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 0.1, max_depth = 30, n_estimators = 100)



ml = [('Linear Regression',LR),
      ('Adabost',Ada),
         ('Random Forest',RF),('xgboost',Xgb)]

r2_train = []
r2_test = []
mae_train = []
mae_test = []
mse_train = []
mse_test = []
rmse_train = []
rmse_test = []
mape_train = []
mape_test = []
models = []

for model in ml:
  for x in range(0,3):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
    model[1].fit(X_train,y_train)

    predict_train = model[1].predict(X_train)
    predict_test = model[1].predict(X_test)

    r2_score_train = r2_score(y_train,predict_train)
    r2_score_test = r2_score(y_test,predict_test)

    mae_score_train = mean_absolute_error(y_train,predict_train)
    mae_score_test = mean_absolute_error(y_test,predict_test)

    mse_score_train = mean_squared_error(y_train,predict_train)
    mse_score_test = mean_squared_error(y_test,predict_test)

    mape_score_train = np.mean(np.abs((y_train - predict_train) / y_train)) * 100
    mape_score_test = np.mean(np.abs((y_test - predict_test) / y_test)) * 100

    rmse_score_train = np.sqrt(mse_score_train)
    rmse_score_test = np.sqrt(mse_score_test)

    r2_train.append(r2_score_train)
    r2_test.append(r2_score_test)
    mae_train.append(mae_score_train)
    mae_test.append(mae_score_test)
    mse_train.append(mse_score_train)
    mse_test.append(mse_score_test)
    rmse_train.append(rmse_score_train)
    rmse_test.append(rmse_score_test)
    mape_train.append(mape_score_train)
    mape_test.append(mape_score_test)
    models.append(model[0])


df_metrics = pd.DataFrame({'model':models,
                           'r2_score_train':r2_train,
                           'r2_score_test':r2_test,
                           'MAE_train':mae_train,
                           'MAE_test':mae_test,
                           'MSE_train':mse_train,
                           'MSE_test':mse_test,
                           'MAPE_train':mape_train,
                           'MAPE_test':mape_test,
                           'RMSE_train':rmse_train,
                           'RMSE_test':rmse_test})

```

# XGBoost Hypertuning

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import mean_squared_error
from scipy.stats import randint, uniform


xgb = XGBRegressor(objective='reg:squarederror', random_state=42)



param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.9, 1.0],
    'colsample_bytree': [0.7, 0.9, 1.0]
}


param_dist = {
    'n_estimators': randint(100, 500),
    'max_depth': randint(3, 10),
    'learning_rate': uniform(0.01, 0.3),
    'subsample': uniform(0.6, 0.4),
    'colsample_bytree': uniform(0.6, 0.4),
    'gamma': uniform(0, 0.5),
    'reg_lambda': uniform(0.5, 1),
}


grid_xgboost = GridSearchCV(estimator=xgb, param_grid=param_grid, cv=5, scoring='neg_root_mean_squared_error', verbose=1, n_jobs=-1)
grid_xgboost.fit(X_train, y_train)


random_search_xgboost = RandomizedSearchCV(estimator=xgb, param_distributions=param_dist, 
                                   n_iter=50, scoring='neg_root_mean_squared_error', 
                                   cv=5, verbose=1, n_jobs=-1, random_state=42)
random_search_xgboost.fit(X_train, y_train)

print("Grid")

print("Best parameters:", grid_xgboost.best_params_)
print("Best score (neg RMSE):", grid_xgboost.best_score_)

print("random search")

print("Best parameters:", random_search_xgboost.best_params_)
print("Best score (neg RMSE):", random_search_xgboost.best_score_)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score


best_model_xgboost = random_search_xgboost.best_estimator_
y_pred_test_xgboost = best_model_xgboost.predict(X_test)
y_pred_train_xgboost = best_model_xgboost.predict(X_train)


# Evaluate
print("Train")

print("MAE:", mean_absolute_error(y_train, y_pred_train_xgboost))
print("R2 Score:", r2_score(y_train, y_pred_train_xgboost))
print("MAPE:", np.mean(np.abs((y_train - y_pred_train_xgboost) / y_train)) * 100)


print("Test")

print("MAE:", mean_absolute_error(y_test, y_pred_test_xgboost))
print("R2 Score:", r2_score(y_test, y_pred_test_xgboost))
print("MAPE:", np.mean(np.abs((y_test - y_pred_test_xgboost) / y_test)) * 100)
```
